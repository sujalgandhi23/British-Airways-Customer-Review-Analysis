```python
for col in ba.select_dtypes(include=['object', 'string']):
    ba[col] = ba[col].str.strip()
```

```python
ba["date"] = pd.to_datetime(ba["date"], dayfirst=True)
ba["date_flown"] = pd.to_datetime(ba["date_flown"], dayfirst=True)
```

```python
ba["traveller_type"] = ba["traveller_type"].fillna("Unknown")
```
